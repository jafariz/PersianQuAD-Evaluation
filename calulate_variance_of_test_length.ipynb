```python
ds = load_dataset(ds_name)
```

```python
import statistics
```

```python
ds
```

```python
train_answer_len = []
for d in ds['train']:
  if len(d['answers']['text']) > 0 :
    train_answer_len.append(len(d['answers']['text'][0]))
  else:
     train_answer_len.append(0)
```

```python
train_answer_len_var = statistics.variance(train_answer_len)
print(f"train_answer_len Variance: {train_answer_len_var}")
```

```python
test_answer_len = []
for d in ds['test']:
  for a in d['answers']['text']:
    test_answer_len.append(len(d['answers']['text'][0]))
```

```python
test_answer_len_var = statistics.variance(test_answer_len)
print(f"test_answer_len Variance: {test_answer_len_var}")
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# 2. Count the frequency of each number
frequencies = Counter(train_answer_len)

# Separate the numbers (keys) and their counts (values)
numbers = list(frequencies.keys())
counts = list(frequencies.values())

# 3. Create the plot
plt.figure(figsize=(8, 5))
plt.bar(numbers, counts, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Number/Rating', fontsize=12)
plt.ylabel('Frequency (Count)', fontsize=12)
plt.title('Frequency of Ratings', fontsize=14, fontweight='bold')

# Ensure x-axis only shows our discrete integers
plt.xticks(numbers)

# Add a clean grid behind the bars
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
#plt.yscale('log')
plt.show()
```

```python
from collections import Counter

# Assuming 'data' is your list of numbers
counter = Counter(train_answer_len)

# Get the top 5 most frequent numbers and their exact counts
top_5 = counter.most_common(5)

print("Top 5 most frequent numbers (Number, Frequency):")
for number, frequency in top_5:
    print(f"Number: {number} appeared {frequency} times")
```
